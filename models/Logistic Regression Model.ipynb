```python
import pandas as pd
import numpy as np
import matplotlib as plt

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("new_train.csv")
test = pd.read_csv("new_test.csv")
validation = pd.read_csv("new_validation.csv")
```

```python
train = train.drop(train.columns[0], axis = 1)
test = test.drop(test.columns[0], axis = 1)
```

```python
train.head()
```

# Logistic Regression Model

```python
train_x = train[train.columns[0:20]]
train_y = train[train.columns[20]]
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(train_x, train_y)
```

```python
class_le = LabelEncoder()
test['States'] = class_le.fit_transform(test['States'].values)
test.dropna()
```

```python
test_x = test[test.columns[0:20]]
test_y = test[test.columns[20]]
```

```python
logisticRegr.predict(test_x)
```

```python
score = logisticRegr.score(test_x, test_y)
print(f"Model achieved the score of: {round(score * 100, 2)} %")
```
